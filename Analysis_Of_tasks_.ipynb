```python
import matplotlib.pyplot as plt
import seaborn as sns

columns_to_group = data.columns[3:-2]

sns.set(style="whitegrid")

for column in columns_to_group:
    counts = data.groupby(column).size().sort_values(ascending=False)

    plt.figure(figsize=(10, 5))
    ax = counts.plot(kind='barh', color=sns.color_palette('Dark2'))

    for i in ax.patches:
        ax.text(i.get_width() + 0.1, i.get_y() + i.get_height() / 2,
                str(int(i.get_width())), fontsize=10, verticalalignment='center')

    plt.title(f'Counts for {column}')
    plt.xlabel('Count')
    plt.ylabel(column)
    plt.gca().spines[['top', 'right']].set_visible(False)

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

columns_to_group = data.columns[3:-2]

sns.set(style="whitegrid")

for column in columns_to_group:
    counts = data.groupby(column).size().sort_values(ascending=False)

    plt.figure(figsize=(10, 5))
    ax = counts.plot(kind='barh', color=sns.color_palette('Dark2'))

    for i in ax.patches:
        ax.text(i.get_width() + 0.1, i.get_y() + i.get_height() / 2,
                str(int(i.get_width())), fontsize=10, verticalalignment='center')

    plt.title(f'Counts for {column}')
    plt.xlabel('Count')
    plt.ylabel(column)

    plt.gca().spines[['top', 'right']].set_visible(False)

    plt.tight_layout()
    plt.savefig(f'counts_plot_{column}.png')

    plt.close()
```

```python
import openpyxl
import pandas as pd

from openpyxl import Workbook

wb = Workbook()
ws = wb.active
ws.title = "Groupby Counts"

columns_to_group = data.columns[3:-2]


for col_idx, column in enumerate(columns_to_group, start=1):

    counts = data.groupby(column).size().reset_index(name='Count')

    ws.cell(row=1, column=col_idx*2 - 1, value=column)
    ws.cell(row=1, column=col_idx*2, value="Count")
    for row_idx, (group_value, count) in enumerate(counts.itertuples(index=False), start=2):
        ws.cell(row=row_idx, column=col_idx*2 - 1, value=group_value)
        ws.cell(row=row_idx, column=col_idx*2, value=count)
wb.save('groupby_counts.xlsx')
```
